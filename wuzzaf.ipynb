```python
import requests
import csv
from bs4 import BeautifulSoup
from itertools import zip_longest
```

```python
job_title = []
company_name = []
location_name = []
time = []
links = []
posted = []
page_num = 0
```

```python
result = requests.get('https://wuzzuf.net/search/jobs/?q=python&a=hpb')
```

```python
src = result.content
```

```python
soup = BeautifulSoup(src,"lxml")
```

```python
job_titles = soup.findAll("h2",{"class":"css-m604qf"})
company_names = soup.findAll("a",{"class":"css-17s97q8"})
locations_names = soup.findAll("span",{"class":"css-5wys0k"})
times = soup.findAll("div",{"class":"css-1lh32fc"})
posted_new = soup.findAll("div",{"class":"css-4c4ojb"})
posted_old = soup.findAll("div",{"class":"css-do6t5g"})
post = [*posted_new,*posted_old]
```

```python
for i in range(len(job_titles)):
    job_title.append(job_titles[i].text)
    links.append(job_titles[i].find("a").attrs['href'])
    company_name.append(company_names[i].text)
    location_name.append(locations_names[i].text)
    time.append(times[i].text)
    posted.append(post[i].text)
```

```python
list_file = [job_title,company_name,location_name,time,links,posted]
exported = zip_longest(*list_file)
```

```python
with open("myfile.csv","w") as file:
    wr = csv.writer(file)
    wr.writerow(["job title","Company name","location name","time","links","posted"])
    wr.writerows(exported)
```
